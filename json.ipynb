```python
response = requests.get("http://jsonplaceholder.typicode.com/todos")

todos= json.loads(response.text)
print(type(todos))
pp(todos)
```

```python
pp(todos[:5]) #slicing will print out first 5 from 0-4
```

```python
completed=[]
for i in todos:
   # print(i)
    if i["completed"]:
        completed.append(i)

pp(completed)
```

```python
user_todos={}

for i in todos:
    if i['completed']:
        if i['userId']in user_todos:
            user_todos[i['userId']] +=1
        else:
            user_todos[i['userId']] =0
print(user_todos)
        
```

```python
todo_counter ={}
for todo in todos:
    if i ['completed']:
        try:
            todo_counter[todo['userId']] +=1
        except KeyError:
            todo_counter[todo['userId']] = 1

print(todo_counter)
```

```python
count_todos={}
for i in todos:
    if i['completed']:
        count_todos[i['userId']]= count_todos.get(i['userId'], 0) +1
print(count_todos)
```

```python
top_users =sorted(user_todos.items(), key=lambda x: x[1], reverse=True)
print(top_users)
```

```python
users= []
max_value= top_users[0][1]
print(max_value)
for x, y in top_users:
    #print('x is ' + str(x))
    #print('y is ' + str(y))
    if y == max_value:
        users.append(x)

print(users)
```
